```python
import talib as ta
import numpy as np
import pandas as pd

data=pd.read_excel('sz50.xlsx', sheetname='600036.XSHG', index_col='datetime')
Series=data.close
array=np.array(Series)
print(type(array))
print(ta.MA(array,10)[-5:])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
stock1=data.close
rolling_mean=stock1.rolling(window=10).mean()

plt.figure(figsize=(15, 7))
stock1.plot()
rolling_mean.plot()
plt.ylabel('Price')
plt.legend(['600036','601318'])
plt.show()
```
